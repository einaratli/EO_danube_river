```python

import os
from dotenv import load_dotenv

load_dotenv()

# Sækjum lyklana úr kerfisbreytunum
COPERNICUS_CLIENT_ID= os.getenv("COPERNICUS_ID")
COPERNICUS_CLIENT_SECRET = os.getenv("COPERNICUS_KEY")

# Copernicus Dataspace API endpoints
# Bættu við 'scope' eða notaðu þennan hlekk ef hann er annar en sá sem þú ert með
AUTH_URL = "https://identity.dataspace.copernicus.eu/auth/realms/CDSE/protocol/openid-connect/token"
SEARCH_URL = "https://catalogue.dataspace.copernicus.eu/odata/v1/Products"
DOWNLOAD_URL = "https://zipper.dataspace.copernicus.eu/odata/v1/Products"

```

```python
def get_access_token(client_id, client_secret):
    """
    Get OAuth2 access token from Copernicus Dataspace using Client Credentials flow.
    
    This is the recommended method for server-to-server authentication and HPC jobs.
    
    Parameters:
    -----------
    client_id : str
        OAuth2 Client ID (starts with 'sh-')
    client_secret : str
        OAuth2 Client Secret
    
    Returns:
    --------
    str : Access token if successful, None otherwise
    """
    data = {
        "grant_type": "client_credentials",
        "client_id": client_id,
        "client_secret": client_secret,
    }
    
    try:
        response = requests.post(AUTH_URL, data=data, timeout=30)
        response.raise_for_status()
        token_data = response.json()
        
        # Extract token and expiration
        access_token = token_data["access_token"]
        expires_in = token_data.get("expires_in", 3600)
        
        print(f"✓ Token obtained (valid for {expires_in//60} minutes)")
        return access_token
        
    except requests.exceptions.HTTPError as e:
        if e.response.status_code == 401:
            print("❌ Authentication failed: Invalid credentials")
            print("   ✗ Check your CLIENT_ID and CLIENT_SECRET")
            print("   ✗ CLIENT_ID should start with 'sh-'")
        else:
            print(f"❌ HTTP {e.response.status_code}: {e.response.text}")
        return None
        
    except Exception as e:
        print(f"❌ Authentication failed: {e}")
        return None

# Get access token
print("Authenticating with Copernicus Dataspace...")
access_token = get_access_token(COPERNICUS_CLIENT_ID, COPERNICUS_CLIENT_SECRET)

if access_token:
    print("✓ Successfully authenticated")
    headers = {"Authorization": f"Bearer {access_token}"}
else:
    print("❌ Authentication failed.")
    print("\nTroubleshooting:")
    print("1. Check environment variables are set:")
    print(f"   COPERNICUS_CLIENT_ID = {COPERNICUS_CLIENT_ID[:15]}...")
    print(f"   COPERNICUS_CLIENT_SECRET = {COPERNICUS_CLIENT_SECRET[:15]}...")
    print("2. Verify credentials in Copernicus Dashboard")
    print("3. See COPERNICUS_SETUP.md for detailed instructions")
    headers = None
```

```python
# Bounding box coordinates fyrir Danube Bend (Ungverjaland)
# Þetta svæði er innan Sentinel-2 Tile: 33TYN
min_lon, min_lat = 18.50, 47.70
max_lon, max_lat = 19.00, 47.95

# Create WKT POLYGON for API query
roi_polygon = f"POLYGON(({min_lon} {min_lat},{max_lon} {min_lat},{max_lon} {max_lat},{min_lon} {max_lat},{min_lon} {min_lat}))"

# Skilgreinum 4 dagsetningar með jöfnu millibili (ca. 2 mánuðir)
# Þessir dagar eru valdir vegna lítillar skýjahulu í tile 33TYN
target_dates = [
    '2018-03-24',  # Vor: Hátt vatnsborð
    '2018-06-02',  # Snemma sumar: Gróður í hámarki
    '2018-08-11',  # Seint sumar: Hitabylgja byrjuð að bíta
    '2018-10-25'   # Haust: Sögulegt lágmark vatnsborðs
]

# Maximum cloud cover percentage
max_cloud_cover = 30

print(f"Search Parameters:")
print(f"  Region: Danube Bend / Dunakanyar (Hungary)")
print(f"  Bounding Box: ({min_lon}, {min_lat}) to ({max_lon}, {max_lat})")
print(f"  Max Cloud Cover: {max_cloud_cover}%")
print(f"\nValdar dagsetningar til greiningar (jöfn millibil):")

for i, date in enumerate(target_dates, 1):
    print(f"  Tímabil {i}: {date}")

print(f"\nNote: Notaðu þessar dagsetningar í API leitina þína til að bera saman")
print(f"breytingar á vatnsflatarmáli (MNDWI) yfir tímabilið.")
```

```python
import requests
from datetime import datetime, timedelta

SEARCH_URL = "https://catalogue.dataspace.copernicus.eu/odata/v1/Products"

def search_sentinel2_strict(start_date, end_date, roi_polygon, max_cloud_cover):
    # Bætum við 'contains(Name, 'T33TYN')' til að þvinga niðurstöður á þetta Tile
    filters = [
        "Collection/Name eq 'SENTINEL-2'",
        "Attributes/OData.CSC.StringAttribute/any(att:att/Name eq 'productType' and att/OData.CSC.StringAttribute/Value eq 'S2MSI2A')",
        f"ContentDate/Start gt {start_date}T00:00:00.000Z",
        f"ContentDate/Start lt {end_date}T23:59:59.999Z",
        "contains(Name, 'T33TYN')", 
        f"OData.CSC.Intersects(area=geography'SRID=4326;{roi_polygon}')",
        f"Attributes/OData.CSC.DoubleAttribute/any(att:att/Name eq 'cloudCover' and att/OData.CSC.DoubleAttribute/Value lt {max_cloud_cover})"
    ]
    
    filter_query = " and ".join(filters)
    params = {"$filter": filter_query, "$orderby": "ContentDate/Start asc", "$top": 10}
    
    try:
        response = requests.get(SEARCH_URL, params=params, timeout=60)
        response.raise_for_status()
        return response.json().get('value', [])
    except Exception as e:
        print(f"❌ Leit mistókst: {e}")
        return []

# --- Uppfærð hnit fyrir 33TYN ---
min_lon, min_lat = 18.50, 47.70  # Hliðrað til vesturs
max_lon, max_lat = 19.00, 47.95
roi_polygon = f"POLYGON(({min_lon} {min_lat},{max_lon} {min_lat},{max_lon} {max_lat},{min_lon} {max_lat},{min_lon} {min_lat}))"

target_dates = ['2018-03-24', '2018-06-02', '2018-08-11', '2018-10-25']
max_cloud_cover = 30
window_days = 14# Rýmkum gluggann aðeins meira til að finna hreinar myndir

all_results = {}

print("Leita að myndum eingöngu í Tile 33TYN...\n")

for date_str in target_dates:
    base_date = datetime.strptime(date_str, '%Y-%m-%d')
    start_search = (base_date - timedelta(days=window_days)).strftime('%Y-%m-%d')
    end_search = (base_date + timedelta(days=window_days)).strftime('%Y-%m-%d')
    
    results = search_sentinel2_strict(start_search, end_search, roi_polygon, max_cloud_cover)
    
    if results:
        best_match = results[0]
        actual_date = best_match['ContentDate']['Start'][:10]
        all_results[date_str] = best_match
        print(f"✓ {date_str} -> Fann 33TYN mynd þann {actual_date}. ID: {best_match['Id']}")
    else:
        print(f"⚠️ {date_str} -> Engin mynd fannst í 33TYN með < 30% ský.")
```

```python
from collections import defaultdict

# Söfnum öllum myndum sem fundust í eina lista úr 'all_results' orðabókinni
products = [p for p in all_results.values() if p is not None]

# Group products by MGRS tile
tiles = defaultdict(list)

for product in products:
    product_name = product.get('Name', '')
    
    # Sentinel-2 nafnasnið: S2B_MSIL2A_20181025T095029_N0209_R079_T33TYN_20181025T135509.SAFE
    # Tile ID er yfirleitt í 6. hluta nafnsins (split með '_')
    parts = product_name.split('_')
    if len(parts) > 5:
        # Hreinsum 'T' framan af ef það er til staðar (t.d. T33TYN -> 33TYN)
        tile_id = parts[5]
        if tile_id.startswith('T'):
            tile_id = tile_id[1:]
    else:
        tile_id = 'Unknown'
        
    tiles[tile_id].append(product)

# Birta niðurstöður
print("Available MGRS Tiles and Acquisition Counts:")
print("=" * 50)

if not tiles:
    print("Engar myndir fundust til að flokka. Athugaðu leitarniðurstöður.")
else:
    # Röðum eftir fjölda mynda á hverju Tile
    sorted_tiles = sorted(tiles.items(), key=lambda x: len(x[1]), reverse=True)
    
    for tile_id, tile_products in sorted_tiles:
        print(f"\nTile {tile_id}: {len(tile_products)} acquisitions")
        
        # Röðum vörum innan hvers Tile eftir dagsetningu
        tile_products.sort(key=lambda x: x.get('ContentDate', {}).get('Start', ''))
        
        for i, product in enumerate(tile_products):
            date = product.get('ContentDate', {}).get('Start', 'N/A')[:10]
            # Stærð í GB
            size_bytes = product.get('ContentLength', 0)
            size_gb = size_bytes / (1024**3) if size_bytes else 0
            
            print(f"  {i+1}. {date} - {size_gb:.2f} GB - ID: {product.get('Id')}")

print("\n" + "=" * 50)

# Ráðlegging byggð á niðurstöðum
if any(len(p) >= 4 for p in tiles.values()):
    print(f"\nRecommendation: Excellent! Þú fannst Tile með öllum 4 tímabilunum.")
else:
    print(f"\nRecommendation: Athugaðu að rýmka tímagluggann ef þig vantar fleiri tímabil á sama Tile.")
```

```python
import os
import requests

# Skilgreinum grunnslóðina fyrir niðurhal í Copernicus
# Uppfærðu þetta í upphafi kóðans
DOWNLOAD_URL = "https://zipper.dataspace.copernicus.eu/odata/v1/Products"

# Þín sérstaka slóð
data_path = "/p/project1/training2600/TeamGudnason/data"

def download_product(product, output_dir, access_token):
    product_id = product['Id']
    product_name = product['Name']
    
    # Búum til möppuna ef hún er ekki til
    os.makedirs(output_dir, exist_ok=True)
    
    output_file = os.path.join(output_dir, f"{product_name}.zip")
    
    if os.path.exists(output_file):
        print(f"⚠ Skrá er þegar til: {product_name}.zip")
        return output_file
    
    # Ath: DOWNLOAD_URL verður að vera skilgreint fyrir utan fallið
    download_url = f"{DOWNLOAD_URL}({product_id})/$value"
    
    headers = {
        "Authorization": f"Bearer {access_token}",
        "Accept": "*/*"  # Hleypa öllu í gegnum audience athugunina
}
    
    try:
        print(f"Sæki: {product_name}")
        print(f"  Áætluð stærð: {product['ContentLength'] / (1024**3):.2f} GB")
        
        with requests.get(download_url, headers=headers, stream=True, timeout=300) as response:
            response.raise_for_status()
            
            total_size = int(response.headers.get('content-length', 0))
            block_size = 8192
            downloaded = 0
            
            with open(output_file, 'wb') as f:
                for chunk in response.iter_content(chunk_size=block_size):
                    if chunk:
                        f.write(chunk)
                        downloaded += len(chunk)
                        
                        # Prenta framvindu á 100 MB fresti
                        if downloaded % (100 * 1024 * 1024) < block_size:
                            progress = (downloaded / total_size) * 100 if total_size > 0 else 0
                            print(f"  Framvinda: {progress:.1f}% ({downloaded / (1024**3):.2f} GB)")
        
        print(f"✓ Niðurhali lokið: {output_file}")
        return output_file
        
    except Exception as e:
        print(f"❌ Niðurhal mistókst: {e}")
        if os.path.exists(output_file):
            os.remove(output_file)
        return None

```

```python
# 1. Ná í nýtt token (passaðu að nota ferskt token)
access_token = get_access_token(COPERNICUS_CLIENT_ID, COPERNICUS_CLIENT_SECRET)

# 2. Skilgreina hvert á að vista
data_path = "/p/project1/training2600/TeamGudnason/data"

if access_token and products:
    print(f"Byrja að hlaða niður {len(products)} myndum...")
    for product in products:
        # Fallið þitt notar DOWNLOAD_URL breytuna úr global scope
        download_product(product, data_path, access_token)
else:
    print("❌ Auðkenning mistókst eða engar vörur fundust.")
```

```python
if access_token and products:
    print(f"Byrja að hlaða niður {len(products)} myndum í {data_path}...")
    
    downloaded_files = []
    
    for product in products:
        # Köllum í fallið þitt með nýja token-inu
        file_path = download_product(product, data_path, access_token)
        
        if file_path:
            downloaded_files.append(file_path)
            
    print(f"\n" + "="*50)
    print(f"✓ Verki lokið! {len(downloaded_files)} myndir eru komnar á áfangastað.")
    print("="*50)
else:
    print("❌ Ekkert hægt að gera. Athugaðu hvort 'access_token' og 'products' listinn séu til staðar.")
```
